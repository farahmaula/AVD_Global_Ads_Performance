**5. CPA**

Pada kolom CPA persentase outliers yaitu 7,7%. Outlier CPA sering memperlihatkan campaign iklan yang performanya kurang, contohnya biaya iklan yang dikeluarkan besar sedangkan konversinya sangat rendah. Nilai yang ekstrem bisa mengganggu analisis performa seluruhnya terutama dalam menghitung rata-rata efisiensi biaya antar campaign. Perlunya metode IQR (Interquartile Range) untuk menangani hal ini.

Metode IQR (Interquartile Range) adalah cara untuk menentukan "pagar pembatas" agar kita bisa membedakan mana data yang masih dianggap wajar dan mana data yang sudah dianggap aneh (outlier). Metode ini bekerja dengan mengabaikan data yang terlalu pendek atau terlalu tinggi, dan hanya fokus pada 50% data di bagian tengah (pusat sebaran data).

Mengapa memilih menggunakan metode IQR? karena batas wajarnya ditentukan berdasarkan karakteristik persebaran data itu sendiri.


```python
Q1 = df['CPA'].quantile(0.25)
Q3 = df['CPA'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['CPA'] = df['CPA'].clip(lower=lower_bound, upper=upper_bound)
```

**6. Revenue**

Persentase outlier pada kolom revenue yaitu 7,05%. Revenue adalah pendapatan kotor yang masuk pada perusahaan. Pada beberapa campaign menghasilkan pendapatan yang lebih besar dibanding lainnya. Dengan melakukan penanganan pada kolom ini bisa membuat perhitungan total pendapatan tidak akurat, oleh sebab itu tidak perlu ditangani.

**7. ROAS**

Pada kolom ROAS persentase outliernya yaitu 6,83%. Adanya outliers pada ROAS dikarenakan campaign mengeluarkan biaya iklan lebih besar dibandingkan pendapatannya. Nilai yang sangat tinggi harus dikendalikan agar tidak terjadinya perubahan yang besar. Maka dari itu perlu metode IQR (Interquartile Range).

Metode adalah cara untuk menentukan "pagar pembatas" agar kita bisa membedakan mana data yang masih dianggap wajar dan mana data yang sudah dianggap aneh (outlier). Metode ini bekerja dengan mengabaikan data yang terlalu pendek atau terlalu tinggi, dan hanya fokus pada 50% data di bagian tengah (pusat sebaran data). Alasan mengapa menggunakan metode ini karena batasannya batasannya ditentukan oleh karakteristik persebaran data itu sendiri.

```python
Q1 = df['ROAS'].quantile(0.25)
Q3 = df['ROAS'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['ROAS'] = df['ROAS'].clip(lower=lower_bound, upper=upper_bound)
```

# **9. Construct Data**

Construct data adalah membuat fitur baru dari data yang sudah ada untuk memperbanyak dalam analisis.

1. Membuat Kolom Baru "profit"

Kolom profit ini digunakan untuk memperbanyak analisis dan untuk mengetahui keuntungan bersih yang didapat dari revenue (pendapatan kotor) dan ad spend (pengeluaran biaya).

```python
df['profit'] = df['revenue'] - df['ad_spend']

def keuntungan_bersih(profit):
    if profit > 0:
        return 'Untung'
    elif profit < 0:
        return 'Rugi'
    else:
        return 'Break Even'

df['profit'] = df['profit'].apply(keuntungan_bersih)
```
